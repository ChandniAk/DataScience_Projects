```python
# Remove non-numeric values from 'streams' column and convert to integer
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df['streams'].dtypes
df['streams']
```

```python
top_streams = df[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
top_streams
```

```python
# Top 10 songs with most streams on Spotify
top_streams = df[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
sns.barplot(x=top_streams['streams'], y=top_streams['track_name'],  palette='viridis')
```

```python
# Calculate average danceability for each year
dance_per_year = df.groupby(by='released_year')['danceability_%'].mean()
dance_per_year
```

```python
sns.lineplot(x=dance_per_year.index, y = dance_per_year.values, palette=None)
```

```python
# Select columns for cross-platform metrics
cross_platform_columns = [
    'in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists', 
    'in_apple_charts', 'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts'
]

# Calculate the correlation matrix
coll = df[cross_platform_columns].corr()
sns.heatmap(coll, annot=True)
```
